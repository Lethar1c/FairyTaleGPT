### Парсинг данных

Перед тем, как обучать GPT, спарсим все сказки из deti-online.com

```python
import requests
import pandas as pd
from pathlib import Path
from bs4 import BeautifulSoup
```

```python
LINK = "https://deti-online.com/skazki"

st_accept = "text/html"

st_useragent = "Mozilla/5.0 (Macintosh; Intel Mac OS X 12_3_1) AppleWebKit/605.1.15 (KHTML, like Gecko) Version/15.4 Safari/605.1.15"

headers = {
   "Accept": st_accept,
   "User-Agent": st_useragent
}
```

```python
req = requests.get(LINK, headers)
src = req.text
```

```python
soup = BeautifulSoup(src)
```

```python
tales_list = soup.select_one("#list")
```

```python
tales_links = tales_list.select("a")
tales_links = list(map(lambda x: x['href'], tales_links))
tales_links
```

```python
for i, tale in enumerate(tales_links):
    req = requests.get("https://deti-online.com" + tale)
    src = req.text
    soap = BeautifulSoup(src)

    file_link = "https://deti-online.com" + soap.select("a[title='Скачать epub']")[0]['href']

    response = requests.get(file_link)

    if response.status_code == 200:
        with open(f"epubs/{i}.epub", 'wb') as file:
            file.write(response.content)

        print(f"Download tale from {file_link} successfully")

    else:
        print(f"Failed to download tale from {file_link}. Status code: {response.status_code}")
```

```python
import os
from tqdm import tqdm

tales_texts = []

for file_name in tqdm(os.listdir("./epubs")):
    content = textract.process("./epubs/"+file_name, encoding="utf-8").decode()
    tales_texts.append(content)
```

```python
for text in tales_texts:
    with open("corpus.txt", "a") as file:
        file.write("\n<BOS>\n")
        file.write(text)
        file.write("\n<EOS>\n")
```
